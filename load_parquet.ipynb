```python
from chroma.db.clickhouse import Clickhouse
from chroma.config import Settings
import numpy as np
import pyarrow.parquet as pq
import time
```

```python
# Setup a DB connection
chroma_settings = Settings()
db = Clickhouse(chroma_settings)

print(db.count())

# Load in the parquet file
file = ".chroma/chroma.parquet"
print("Loading parquet file: ", file)
py = pq.read_table(file)
df = py.to_pandas()
print("Number of records: ", len(df))

# Set model stpace
model_space = "default_scope"
```

```python
BATCH_SIZE = 10_000
data_length = len(df)

for i in range(0, data_length, BATCH_SIZE):
    # Time the following code
    start = time.time()    
    batch = df[i:i+BATCH_SIZE]

    embedding = np.array(batch['embedding_data'].tolist()).tolist()
    input_uri = batch['input_uri'].tolist()
    label_class = batch['label_category_name'].tolist()
    inference_class = batch['inference_category_name'].tolist()
    dataset = batch['dataset'].tolist()

    db.add(model_space=[model_space]*len(batch), embedding=embedding, input_uri=input_uri, label_class=label_class, inference_class=inference_class, dataset=dataset)
    end = time.time()
    print("Time to process BATCH_SIZE rows: " + '{0:.2f}'.format((end - start)) + "s, records loaded: " + str(i))

print(db.count())
    
```
